```python
import numpy as np
import math
import matplotlib.pyplot as plt 
from scipy.stats import rayleigh
```

```python
n = 10000
sgm = 4
```

```python
#1
q = np.random.rayleigh(sgm, size = n)
```

```python
#2.1
print("min = ", min(q)) # минимальное значение
print("max = ", max(q)) # максимальное значение
print("выборочное среднее = ", np.mean(q)) # выборочное среднее
print("смещенная дисперсия = ", np.var(q)) # выборочная смещенная (не исправленная) дисперсия
print("несмещенная дисперсия = ", np.var(q,ddof=1)) # выборочная несмещенная (исправленная) дисперсия
```

```python
def analysis23(n):
    m = []
    d = []
    for i in range(1000):
        c = np.random.rayleigh(sgm, size = n)
        m.append(np.mean(c))#Выборочное среднее
        d.append(np.var(c,ddof=1))#Выборочная несмещенная (исправленная) дисперсия
    return m,d

#2.2
Mx = sgm * math.sqrt(math.pi / 2)
Dx = (4 - math.pi) / 2 * (sgm**2)
print("матожидание", Mx)
print("дисперсия", Dx)
```

```python
#2.3, n = 10
m, d = analysis23(10)
plt.plot(range(0,len(m)),m, color='green',linewidth=3) # рисуем график зависимости выборочного среденего от номера реализации
plt.axhline(y=Mx, color='yellow')# Добавление горизонтальной линии на уровне y=Mx (значение математического ожидания)
```

```python
plt.plot(range(0,len(d)),d, color='green',linewidth=3) # рисуем график зависимости выборочного среденего от номера реализации
plt.axhline(y=Dx, color='red')# Добавление горизонтальной линии на уровне y=Dx (значение дисперсии)
```

```python
#3
xmin = np.min(q)
xmax = np.max(q)
print("min = ", xmin)
print("max = ", xmax)

r = int(np.ceil(1 + np.log2(n)))
print("количество интервалов = ", r)

h = (xmax - xmin) / r
print("ширина интервалов = ", h)

z = np.linspace(xmin, xmax, r + 1)
print("границы интервалов:")
for i, bound in enumerate(z):
    print("z[",i,"] = ", bound)

counter, bins = np.histogram(q, bins=z)
print("количество значений в интервалах ui:", counter)

relFreq = counter / (h * n) # отн. частота

# гистограмма
plt.figure(figsize=(10, 6))
plt.bar(z[:-1], relFreq, width=h)

# теор. кривая
x = np.linspace(0, xmax * 1.1, 500)
xL = rayleigh.pdf(x, scale=sgm)
plt.plot(x, xL, 'r-', linewidth=2)
```

```python
#4

mid = (bins[:-1] + bins[1:]) / 2 #середины интервалов

groupMean = np.sum(mid * counter) / n
print("Выборочное среднее по группированным данным = ", groupMean)

groupVar = np.sum(counter * (mid - groupMean)**2) / (n - 1)
print("Исправленная дисперсия по группированным данным = ", groupVar)
 
#из пункта 2.1 (негруппированные данные)
print("Выборочное среднее = ", np.mean(q))
print("Выборочная несмещенная дисперсия = ", np.var(q,ddof=1))
```

```python
#5
#эмпирическая функция по группированным 
Y = counter.cumsum() / n #накопленные частоты

for i in range(len(Y)):
    plt.plot([z[i], z[i+1]], [Y[i], Y[i]], color='green', linewidth=2.5) #строим


# теоретическая функция, для распределения Рэлея: F(x) = 1 - exp(-x^2/(2*sgm^2))
xT = np.linspace(0, xmax * 1.1, 1000)
fXT = rayleigh.cdf(xT, scale=sgm)
plt.plot(xT, fXT, 'r')
```

```python
#по группированным
sortedQ = np.sort(q) #сортированная выборка
fx = np.arange(1, n + 1) / n # массив элементов <= x, деленных на n
plt.step(sortedQ, fx, linewidth=5) 

plt.plot(xT, fXT, 'r')
```
